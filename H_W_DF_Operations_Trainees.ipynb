#### Create a new dataframe with a column called HV Ratio that is the ratio of the High Price versus volume of stock traded for a day.

```python
new_df = df.withColumn('HV Ratio', df.High / df.Volume)
hv_ratio_df = new_df.select('HV Ratio')
hv_ratio_df.show()
```

#### What day had the Peak High in Price?

```python
peak_high_day = df.orderBy(df['High'].desc()).select('Date').first()
print(peak_high_day['Date'])
```

#### What is the mean of the Close column?

```python
from pyspark.sql.functions import mean

mean_df = df.select(mean(df['Close']).alias('mean_close'))
mean_df.show()
```

#### What is the max and min of the Volume column?

```python
max_min_close = df.select(max(df['Close']).alias('max_close'), min(df['Close']).alias('min_close'))

max_min_close.show()
```

#### How many days was the Close lower than 60 dollars?

```python
num_days = df.filter(df['Close'] < 60).count()

print(num_days)
```

#### What percentage of the time was the High greater than 80 dollars ?
#### In other words, (Number of Days High>80)/(Total Days in the dataset)

```python
num_days_high_gt_80 = df.filter(df['High'] > 80).count()

total_days = df.count()

percentage = (num_days_high_gt_80 / total_days) * 100

print(percentage)
```

#### What is the Pearson correlation between High and Volume?

```python
df.createOrReplaceTempView("stock_data")

correlation_df = spark.sql("SELECT corr(High, Volume) as correlation FROM stock_data")

correlation_df.show()
```

#### What is the max High per year?

```python
df.createOrReplaceTempView("stock_data")

max_high_per_year = spark.sql("""
    SELECT YEAR(Date) as Year, MAX(High) as Max_High
    FROM stock_data
    GROUP BY YEAR(Date)
""")

max_high_per_year.show()
```

#### What is the average Close for each Calendar Month?
#### In other words, across all the years, what is the average Close price for Jan,Feb, Mar, etc... Your result will have a value for each of these months.

```python
avg_close_per_month = spark.sql("""
    SELECT MONTH(Date) as Month, AVG(Close) as Avg_Close
    FROM stock_data
    GROUP BY MONTH(Date)
    ORDER BY MONTH(Date)
""")

avg_close_per_month.show()
```

### Group by "Name" column and using sum function on "salary" column

```python
group_by_Name = dftest03.groupBy('Name').sum('salary')

group_by_Name.show()
```

### Group by "Name" column and using avg function on "salary" column

```python
group_by_Name_Avg = dftest03.groupBy('Name').avg('salary')

group_by_Name_Avg.show()
```

### Group by "Departments" column and using sum function on "salary" column

```python
group_by_Department_Sum = dftest03.groupBy('Departments').sum('salary')

group_by_Department_Sum.show()
```

### Group by "Departments" column and using mean function on "salary" column:

```python
group_by_Department_Mean = dftest03.groupBy('Departments').mean('salary')

group_by_Department_Mean.show()
```

Group by "Departments" column and using count function on "Departments" column:

```python
group_by_Department_Count = dftest03.groupBy('Departments').count()

group_by_Department_Count.show()
```

### Apply agg to using sum function get the total of salaries

```python
from pyspark.sql import functions as F

sum_Salarys = dftest03.agg(F.sum('salary'))

sum_Salarys.show()
```
